```python
import os
import math
import torch
import requests
import numpy as np
from PIL import Image
from tqdm import tqdm  # 导入 tqdm 库
from abc import abstractmethod
import matplotlib.pyplot as plt
from torchvision import transforms
import torch.nn.functional as F
import torch_optimizer as optim
from torchvision import datasets, transforms
from torch.utils.data import Dataset, DataLoader


from DDIM_C import GaussianDiffusion
from ldm.models.autoencoder import VQModelInterface
from ldm.modules.diffusionmodules.openaimodel import UNetModel

# 检查是否有GPU可用
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

if torch.cuda.is_available():
    device_name = torch.cuda.get_device_name(0)
    print(f"GPU Name: {device_name}")
else:
    print("No GPU available, using CPU.")
```

```python
#Step 1 : 定义数据加载器
# 数据转换
transform2 = transforms.Compose([
    transforms.ToTensor(),  # 将图像转换为张量
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # 将像素值限制在-1到1之间
])

class PalmDataset(Dataset):
    def __init__(self, palm_dir, label_dir, transform=None):
        self.palm_dir = palm_dir
        self.label_dir = label_dir
        self.transform = transform
        # 过滤掉非图像文件和目录
        self.palm_images = sorted([f for f in os.listdir(palm_dir) 
                                   if os.path.isfile(os.path.join(palm_dir, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg'))])
        self.label_images = sorted([f for f in os.listdir(label_dir) 
                                    if os.path.isfile(os.path.join(label_dir, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg'))])


    def __len__(self):
        return len(self.palm_images)

    def __getitem__(self, idx):
        palm_path = os.path.join(self.palm_dir, self.palm_images[idx])
        label_path = os.path.join(self.label_dir, self.label_images[idx])
        palm_image = Image.open(palm_path).convert("RGB")
        label_image = Image.open(label_path).convert("RGB")
        # 调整图像大小，确保所有图像大小一致
        palm_image = palm_image.resize((256, 256))
        label_image = label_image.resize((64, 64))

        if self.transform:
            palm_image = self.transform(palm_image)
            label_image = self.transform(label_image)

        return palm_image, label_image


# 定义real_image_folder路径
real_image_folder = '/root/onethingai-fs/paired_datasets/A'
bezier_image_folder='/root/onethingai-fs/paired_datasets/label_E'

# 创建数据集和数据加载器
dataset = PalmDataset(real_image_folder, bezier_image_folder, transform=transform2)

def custom_collate_fn(batch):
    palm_images = torch.stack([item[0] for item in batch])
    label_images = torch.stack([item[1] for item in batch])
    return palm_images, label_images

train_loader = DataLoader(dataset, 
                          batch_size=4, 
                          shuffle=True, 
                          num_workers=8, 
                          pin_memory=True, 
                          collate_fn=custom_collate_fn)
```

```python
#Step 2 : 实例化vqmodel
# 加载模型参数
ckpt_path = 'vqmodel_checkpoint.ckpt'
checkpoint = torch.load(ckpt_path, map_location=device, weights_only=True)

# 简化版VQModel的类
class VQModel(torch.nn.Module):
    def __init__(self, ddconfig, embed_dim=3,n_embed=8192):
        super().__init__()
        from taming.modules.diffusionmodules.model import Encoder, Decoder
        from taming.modules.vqvae.quantize import VectorQuantizer2 as VectorQuantizer
        self.encoder = Encoder(**ddconfig)
        self.decoder = Decoder(**ddconfig)
        self.quantize = VectorQuantizer(n_embed, embed_dim, beta=0.25,
                                        remap=None,
                                        sane_index_shape=False)
        self.quant_conv = torch.nn.Conv2d(ddconfig["z_channels"], embed_dim, 1)
        self.post_quant_conv = torch.nn.Conv2d(embed_dim, ddconfig["z_channels"], 1)

    def encode(self, x):  # 修改方法名称，避免与属性名冲突
        h = self.quant_conv(self.encoder(x))
        return h

    def decode(self, x, force_not_quantize=False):  # 修改方法名称，避免与属性名冲突
        if not force_not_quantize:
            quant, emb_loss, info = self.quantize(x)
        else:
            quant = x
        dec = self.decoder(self.post_quant_conv(quant))
        return dec

# 初始化简化版模型
vq_model = VQModel(  # 实例化对象的名称改为小写以避免与类名混淆
    ddconfig={
        'double_z': False,
        'z_channels': 3,
        'resolution': 256,
        'in_channels': 3,
        'out_ch': 3,
        'ch': 128,
        'ch_mult': [1, 2, 4],
        'num_res_blocks': 2,
        'attn_resolutions': [],
        'dropout': 0.0
    },
    embed_dim=3,
    n_embed=8192
)

# 加载权重
# 过滤掉与模型不相关的参数
if 'model_state_dict' in checkpoint:
    state_dict = checkpoint['model_state_dict']
elif 'state_dict' in checkpoint:
    state_dict = checkpoint['state_dict']
else:
    state_dict = checkpoint  # 如果checkpoint文件本身就是状态字典

filtered_state_dict = {k: v for k, v in state_dict.items() if k in vq_model.state_dict()}
vq_model.load_state_dict(filtered_state_dict, strict=False)

vq_model = vq_model.to(device)
for param in vq_model.parameters():
    param.requires_grad = False
# 设置模型为评估模式   
vq_model.eval()
```

```python
#Step 3 : 实例化unet
unet_config = {
    "image_size": 64,
    "in_channels": 6,
    "out_channels": 3,
    "model_channels": 224,
    "attention_resolutions": [8, 4, 2],
    "num_res_blocks": 2,
    "channel_mult": [1, 2, 3, 4],
    "num_head_channels": 32,
}

# 实例化UNetModel
unet_model = UNetModel(**unet_config)

ckpt_path_u = 'ddim_c.ckpt'
checkpoint_u = torch.load(ckpt_path_u, map_location=device, weights_only=True)

if 'model_state_dict' in checkpoint_u:
    state_dict = checkpoint_u['model_state_dict']
elif 'state_dict' in checkpoint_u:
    state_dict = checkpoint_u['state_dict']
else:
    state_dict = checkpoint_u  # 如果checkpoint文件本身就是状态字典

filtered_state_dict = {k: v for k, v in state_dict.items() if k in unet_model.state_dict()}
unet_model.load_state_dict(filtered_state_dict, strict=False)

unet_model.to(device)
unet_model.eval()

test_input = torch.randn(4, 6, 64, 64).to(device)  # 生成与配置相匹配的随机张量
test_output = unet_model(test_input, timesteps=torch.tensor([0, 1, 2, 3], device=device))
print("Output shape:", test_output.shape)
```

```python
#Step 4 : 实例化GaussianDiffusion
timesteps = 1000
gaussian_diffusion = GaussianDiffusion(timesteps=timesteps,
        beta_schedule='linear',
        linear_start = 0.0015,
        linear_end= 0.0155)
```

```python
#Step 5 : ddpm采样，并可视化

with torch.no_grad():
    # 获取一个 batch
    _, bezier = next(iter(train_loader))
    bezier = bezier.to(device)
    # 打印 batch 的形状以验证
    print(f"Bezier Batch Images Shape: {bezier.shape}")
    

# Sample new images
sample_tensor = gaussian_diffusion.sample(unet_model, image_size=64, batch_size=4, channels=3, c=bezier)

# 取最后一个时间步的数据
sample_tensor = sample_tensor[-1]  # 现在 sample_tensor 是 [4, 3, 64, 64] 的形状

# 将 numpy.ndarray 转换为 PyTorch Tensor
sample_tensor = torch.tensor(sample_tensor)

# 将 Tensor 移动到 vq_model 所在的设备
device = next(vq_model.parameters()).device
sample_tensor = sample_tensor.to(device)

# 解码图像
generated_images = vq_model.decode(sample_tensor)

# 假设 generated_images 是 (4, 3, 256, 256) 的张量
generated_imgs = generated_images.permute(0, 2, 3, 1)  # 变为 (4, 256, 256, 3)
generated_imgs = generated_imgs.cpu().numpy()  # 将张量转换为 NumPy 数组

# 将 Bezier 图像转换为适合显示的格式
bezier_imgs = bezier.permute(0, 2, 3, 1).cpu().numpy()  # 变为 (4, 64, 64, 3)

# 将 [-1, 1] 的像素值范围转换为 [0, 255]
generated_imgs = (generated_imgs + 1.0) * 255 / 2
bezier_imgs = (bezier_imgs + 1.0) * 255 / 2

# 可视化 Bezier 图像
fig, axs = plt.subplots(2, 2, figsize=(12*4, 12*4), constrained_layout=True)
for i in range(4):
    ax = axs[i // 2, i % 2]
    ax.imshow(bezier_imgs[i].astype('uint8'))  # 确保像素值为 uint8 类型
    ax.set_title(f"Bezier Image {i+1}")
    ax.axis("off")
plt.show()

# 可视化生成的图像
fig, axs = plt.subplots(2, 2, figsize=(12*4, 12*4), constrained_layout=True)
for i in range(4):
    ax = axs[i // 2, i % 2]
    ax.imshow(generated_imgs[i].astype('uint8'))  # 确保像素值为 uint8 类型
    ax.set_title(f"DDPM1000 Generated Image {i+1}")
    ax.axis("off")
plt.show()
```

```python
#Step 6 : ddim采样，并可视化

# Sample new images
sample_tensor = gaussian_diffusion.ddim_sample(unet_model, image_size=64, batch_size=4, channels=3, ddim_timesteps=200,cd=bezier)

# 取最后一个时间步的数据
sample_tensor = sample_tensor  # 现在 sample_tensor 是 [4, 3, 64, 64] 的形状

# 将 numpy.ndarray 转换为 PyTorch Tensor
sample_tensor = torch.tensor(sample_tensor)
print(sample_tensor.shape)  # 输出形状以确认

# 将 Tensor 移动到 vq_model 所在的设备
device = next(vq_model.parameters()).device
sample_tensor = sample_tensor.to(device)

# 解码图像
generated_images = vq_model.decode(sample_tensor)


# 假设 generated_images 是 (4, 3, 64, 64) 的张量
imgs = generated_images.permute(0, 2, 3, 1)  # 变为 (4, 256, 256, 3)

imgs = imgs.cpu().numpy()  # 将张量转换为 NumPy 数组
# 将 [-1, 1] 的像素值范围转换为 [0, 255]
imgs = (imgs + 1.0) * 255 / 2

# 可视化生成的图像
fig, axs = plt.subplots(2, 2, figsize=(12*4, 12*4), constrained_layout=True)
for i in range(4):
    ax = axs[i // 2, i % 2]
    ax.imshow(imgs[i].astype('uint8'))  # 确保像素值为 uint8 类型
    ax.set_title(f"DDIM200 Generated Image {i+1}")
    ax.axis("off")
plt.show()
```
